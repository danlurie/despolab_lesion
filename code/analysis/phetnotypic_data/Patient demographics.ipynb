```python
males = patient_demos[patient_demos['sex'] == 'M']
```

```python
females = patient_demos[patient_demos['sex'] == 'F']
```

```python
plt.style.use('seaborn-ticks')
fig, ax = plt.subplots(nrows=1)
g = ax.hist([females.age_at_scan, males.age_at_scan], 14, stacked=True, 
            rwidth=1, label=['Female (22)', 'Male (64)'])
plt.xlim(15, 85)
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.set_xlabel('Age at Scan', fontsize=16)
ax.set_ylabel('Number of Patients', fontsize=16)
fig.suptitle('Patient Age', fontsize=18)
ax.tick_params(labelsize=14)
plt.legend(loc=2, fontsize=14)
```
